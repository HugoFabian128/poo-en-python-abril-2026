# lenguaje python, crear la clase ¨animal¨con los atributos ¨nombre¨ del animal, ¨raza¨ crea el contructor animal, y las funciones ladrar(), correr(), crear los objetos pitbull, chihuahua, nivel principiante, no es interactivo, proporciona tu los datos

```python
# Definición de la clase Animal
class Animal:
    # El constructor: aquí es donde nacen los atributos del objeto
    def __init__(self, nombre, raza):
        self.nombre = nombre
        self.raza = raza

    # Función para que el animal ladre
    def ladrar(self):
        print(f"{self.nombre} dice: ¡Guau, guau! 🐱‍🐉")

    # Función para que el animal corra
    def correr(self):
        print(f"{self.nombre} está corriendo velozmente. 🐱‍🏍")

    def Morder(self):
        print(f"{self.nombre} está mordiendo a la vecina  🤷‍♀️")

# --- Creación de los objetos (Instanciación) ---

# Creamos el objeto 'pitbull' con datos específicos
pitbull = Animal("Rocco", "Pitbull")

# Creamos el objeto 'chihuahua' con sus datos
chihuahua = Animal("Chispa", "Chihuahua")
# creamos el objeto perro callejero con datos
perrocallejero = Animal("Boby", "chafa")

# --- Uso de los objetos ---

print(f"Datos del primer perro: {pitbull.nombre}, Raza: {pitbull.raza}")
pitbull.ladrar()
pitbull.correr()
chihuahua.Morder()

print("-" * 20)

print(f"Datos del segundo perro: {chihuahua.nombre}, Raza: {chihuahua.raza}")
chihuahua.ladrar()
chihuahua.correr()
chihuahua.Morder()

print("-" * 20)

print(f"Datos del tercer perro: {perrocallejero.nombre}, Raza: {perrocallejero.raza}")
perrocallejero.ladrar()
perrocallejero.correr()
chihuahua.Morder()

print(" Programa creado por Hugo Fabian NL = 15")
```
